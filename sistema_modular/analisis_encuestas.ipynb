# Análisis de Encuestas de Comida Rápida


## Módulo: Lector

```python
import csv

def leer_datos(nombre_archivo):
    with open(nombre_archivo, mode="r", encoding="utf-8") as f:
        lector = csv.DictReader(f)
        return list(lector)
```

## Módulo: Modelos

```python
class Encuestado:
    def __init__(self, datos_diccionario):
        self.id = datos_diccionario.get("id")
        self.comida_preferida = datos_diccionario.get("comida_preferida")
        self.frecuencia_consumo = datos_diccionario.get("frecuencia_consumo")
        self.gasto_promedio = float(datos_diccionario.get("gasto_promedio", 0))
        self.satisfaccion_producto = int(datos_diccionario.get("satisfaccion_producto", 0))
        self.satisfaccion_servicio = int(datos_diccionario.get("satisfaccion_servicio", 0))
        self.tiempo_entrega = datos_diccionario.get("tiempo_entrega")
        self.precio_percepcion = datos_diccionario.get("precio_percepcion")
        self.recomendaria = int(datos_diccionario.get("recomendaria_nps", 0))
        self.volveria_comprar = datos_diccionario.get("volveria_comprar")
        self.calificacion_general = int(datos_diccionario.get("calificacion_general", 0))
```

## Módulo: Reportes A (Reportes 1–5 — Diego)

```python
def total_encuestados(encuestados):
    total = len(encuestados)
    print(f"Reporte 1 - Total de encuestados: {total}")


def comida_mas_preferida(encuestados):
    conteo = {}
    for e in encuestados:
        comida = e.comida_preferida
        conteo[comida] = conteo.get(comida, 0) + 1
    mas_preferida = max(conteo, key=lambda c: conteo[c])
    print(f"Reporte 2 - Comida rápida más preferida: {mas_preferida} ({conteo[mas_preferida]} votos)")


def encuestados_por_comida(encuestados):
    conteo = {}
    for e in encuestados:
        comida = e.comida_preferida
        conteo[comida] = conteo.get(comida, 0) + 1
    total = len(encuestados)
    print("\nReporte 3 - Cantidad de encuestados por tipo de comida:")
    for comida, cantidad in conteo.items():
        porcentaje = (cantidad / total) * 100
        print(f"  {comida}: {cantidad} ({porcentaje:.1f}%)")


def frecuencia_consumo_comun(encuestados):
    conteo = {}
    for e in encuestados:
        frecuencia = e.frecuencia_consumo
        conteo[frecuencia] = conteo.get(frecuencia, 0) + 1
    mas_comun = max(conteo, key=lambda f: conteo[f])
    print(f"Reporte 4 - Frecuencia de consumo más común: {mas_comun} ({conteo[mas_comun]} personas)")


def promedio_gasto_general(encuestados):
    total_gasto = sum(e.gasto_promedio for e in encuestados)
    promedio = total_gasto / len(encuestados)
    print(f"Reporte 5 - Promedio de gasto general: ${promedio:.2f}")
```

## Módulo: Reportes B (Reportes 6–10 — Daniella)

```python
def promedio(lista):
    return sum(lista) / len(lista) if lista else 0


def agrupar_por(encuestados, atributo):
    grupos = {}
    for e in encuestados:
        valor = getattr(e, atributo)
        if valor not in grupos:
            grupos[valor] = []
        grupos[valor].append(e)
    return grupos


def promedio_gasto_por_comida(encuestados):
    grupos = agrupar_por(encuestados, "comida_preferida")
    print("\nReporte 6 - Promedio de gasto por comida:")
    resultados = {comida: promedio([e.gasto_promedio for e in lista]) for comida, lista in grupos.items()}
    for comida, valor in sorted(resultados.items(), key=lambda x: x[1], reverse=True):
        print(f"  {comida}: {valor:.2f}")
    print("El mayor gasto indica mayor consumo y preferencia del cliente.")


def satisfaccion_producto(encuestados):
    valores = [e.satisfaccion_producto for e in encuestados]
    print(f"\nReporte 7 - Satisfacción producto: {promedio(valores):.2f}")
    print("Refleja la calidad percibida del producto.")


def satisfaccion_servicio(encuestados):
    valores = [e.satisfaccion_servicio for e in encuestados]
    print(f"\nReporte 8 - Satisfacción servicio: {promedio(valores):.2f}")
    print("Muestra la calidad de atención al cliente.")


def calificacion_general(encuestados):
    valores = [e.calificacion_general for e in encuestados]
    print(f"\nReporte 9 - Calificación general: {promedio(valores):.2f}")
    print("Resume la experiencia general del cliente.")


def distribucion_precios(encuestados):
    conteo = agrupar_por(encuestados, "precio_percepcion")
    total = len(encuestados)
    print("\nReporte 10 - Distribución de percepción de precios:")
    for categoria, lista in conteo.items():
        porcentaje = (len(lista) / total) * 100
        print(f"  {categoria}: {len(lista)} ({porcentaje:.1f}%)")
    print("Muestra cómo los clientes perciben los precios del restaurante.")
```

## Módulo: Reportes C (Reportes 11–15)

```python
def distribucion_tiempo_entrega(encuestados):
    conteo = {}
    for e in encuestados:
        tiempo = e.tiempo_entrega
        conteo[tiempo] = conteo.get(tiempo, 0) + 1
    total = len(encuestados)
    print("\nReporte 11 - Distribución de tiempo de entrega:")
    for categoria, cantidad in conteo.items():
        porcentaje = (cantidad / total) * 100
        print(f"  {categoria}: {cantidad} ({porcentaje:.1f}%)")


def porcentaje_volveria_comprar(encuestados):
    si = sum(1 for e in encuestados if e.volveria_comprar.strip().lower() in ["sí", "si"])
    total = len(encuestados)
    porcentaje = (si / total) * 100
    print(f"\nReporte 12 - Porcentaje que volvería a comprar: {porcentaje:.1f}%")


def nps_general(encuestados):
    promotores = sum(1 for e in encuestados if e.recomendaria >= 9)
    detractores = sum(1 for e in encuestados if e.recomendaria <= 6)
    pasivos = len(encuestados) - promotores - detractores
    total = len(encuestados)
    nps = ((promotores - detractores) / total) * 100
    print(f"\nReporte 13 - NPS General: {nps:.1f}")
    print(f"Promotores: {promotores} | Pasivos: {pasivos} | Detractores: {detractores}")


def nps_por_comida(encuestados):
    datos = {}
    for e in encuestados:
        comida = e.comida_preferida
        if comida not in datos:
            datos[comida] = {"promotores": 0, "detractores": 0, "total": 0}
        datos[comida]["total"] += 1
        if e.recomendaria >= 9:
            datos[comida]["promotores"] += 1
        elif e.recomendaria <= 6:
            datos[comida]["detractores"] += 1
    print("\nReporte 15 - NPS por tipo de comida:")
    for comida, valores in datos.items():
        nps = ((valores["promotores"] - valores["detractores"]) / valores["total"]) * 100
        print(f"  {comida}: {abs(nps):.1f}")
```

## Módulo: Reportes E (Reportes 16–20)

```python
def _promedios_satisfaccion_por_comida(encuestados):
    suma, conteo = {}, {}
    for e in encuestados:
        comida = e.comida_preferida
        suma[comida] = suma.get(comida, 0) + e.satisfaccion_producto
        conteo[comida] = conteo.get(comida, 0) + 1
    return {comida: suma[comida] / conteo[comida] for comida in suma}


def comida_mejor_satisfaccion(encuestados):
    promedios = _promedios_satisfaccion_por_comida(encuestados)
    mejor = max(promedios, key=lambda c: promedios[c])
    print(f"Reporte 16 - Comida con mejor satisfacción: {mejor} ({promedios[mejor]:.2f})")


def comida_peor_satisfaccion(encuestados):
    promedios = _promedios_satisfaccion_por_comida(encuestados)
    peor = min(promedios, key=lambda c: promedios[c])
    print(f"Reporte 17 - Comida con menor satisfacción: {peor} ({promedios[peor]:.2f})")


def relacion_precio_recomendacion(encuestados):
    datos = {}
    for e in encuestados:
        precio = e.precio_percepcion
        if precio not in datos:
            datos[precio] = []
        datos[precio].append(e.recomendaria)
    print("\nReporte 18 - Relación entre precio y recomendación:")
    for precio, lista in datos.items():
        print(f"  {precio}: promedio recomendación = {sum(lista)/len(lista):.2f}")


def relacion_tiempo_satisfaccion(encuestados):
    datos = {}
    for e in encuestados:
        tiempo = e.tiempo_entrega
        if tiempo not in datos:
            datos[tiempo] = []
        datos[tiempo].append(e.satisfaccion_producto)
    print("\nReporte 19 - Relación entre tiempo de entrega y satisfacción:")
    for tiempo, lista in datos.items():
        print(f"  {tiempo}: satisfacción promedio = {sum(lista)/len(lista):.2f}")


def perfil_predominante(encuestados):
    conteo_comida, conteo_frecuencia = {}, {}
    for e in encuestados:
        conteo_comida[e.comida_preferida] = conteo_comida.get(e.comida_preferida, 0) + 1
        conteo_frecuencia[e.frecuencia_consumo] = conteo_frecuencia.get(e.frecuencia_consumo, 0) + 1
    comida_top = max(conteo_comida, key=lambda c: conteo_comida[c])
    frecuencia_top = max(conteo_frecuencia, key=lambda f: conteo_frecuencia[f])
    print("\nReporte 20 - Perfil predominante del consumidor:")
    print(f"  Comida favorita: {comida_top}")
    print(f"  Frecuencia de consumo: {frecuencia_top}")
```

## Clase AnalizarEncuestados

```python
class AnalizarEncuestados:
    def __init__(self, lista_objetos):
        self.encuestados = lista_objetos

    def total_encuestados(self):             total_encuestados(self.encuestados)
    def comida_mas_preferida(self):          comida_mas_preferida(self.encuestados)
    def encuestados_por_comida(self):        encuestados_por_comida(self.encuestados)
    def frecuencia_consumo_comun(self):      frecuencia_consumo_comun(self.encuestados)
    def promedio_gasto_general(self):        promedio_gasto_general(self.encuestados)
    def promedio_gasto_por_comida(self):     promedio_gasto_por_comida(self.encuestados)
    def satisfaccion_producto(self):         satisfaccion_producto(self.encuestados)
    def satisfaccion_servicio(self):         satisfaccion_servicio(self.encuestados)
    def calificacion_general(self):          calificacion_general(self.encuestados)
    def distribucion_precios(self):          distribucion_precios(self.encuestados)
    def distribucion_tiempo_entrega(self):   distribucion_tiempo_entrega(self.encuestados)
    def porcentaje_volveria_comprar(self):   porcentaje_volveria_comprar(self.encuestados)
    def nps_general(self):                   nps_general(self.encuestados)
    def nps_por_comida(self):                nps_por_comida(self.encuestados)
    def comida_mejor_satisfaccion(self):     comida_mejor_satisfaccion(self.encuestados)
    def comida_peor_satisfaccion(self):      comida_peor_satisfaccion(self.encuestados)
    def relacion_precio_recomendacion(self): relacion_precio_recomendacion(self.encuestados)
    def relacion_tiempo_satisfaccion(self):  relacion_tiempo_satisfaccion(self.encuestados)
    def perfil_predominante(self):           perfil_predominante(self.encuestados)
```
