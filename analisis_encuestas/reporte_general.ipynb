# Reporte General de Encuesta de Ingeniería

Este notebook consolida los 20 análisis realizados sobre el archivo `encuesta_ingenieria_10000_respuestas.csv`. Cada sección lee la columna correspondiente y produce su conteo.

```python
import csv

CSV = "encuesta_ingenieria_10000_respuestas.csv"

def leer_csv(ruta):
    """Devuelve todas las filas del CSV sin encabezado."""
    with open(ruta, "r", encoding="utf-8") as f:
        lector = csv.reader(f)
        next(lector)
        return list(lector)
```

---
## Reporte 1 — Total de encuestados
**Columna:** índice 0 (ID del encuestado).  
Se cuenta cada fila que tenga un ID no vacío.

```python
filas = leer_csv(CSV)

total_encuestados = sum(1 for fila in filas if fila[0])

print(f"Total de encuestados: {total_encuestados}")
```

---
## Reporte 2 — Estudiantes por carrera
**Columna:** índice 1 (carrera).  
Se agrupa por las tres ingenierías disponibles.

```python
carreras = ["Ingeniería Eléctrica", "Ingeniería de Sistemas", "Ingeniería Electrónica"]
conteo_carreras = {c: 0 for c in carreras}

for fila in filas:
    carrera = fila[1].strip()
    if carrera in conteo_carreras:
        conteo_carreras[carrera] += 1

for carrera, total in conteo_carreras.items():
    print(f"{carrera}: {total}")
```

---
## Reporte 3 — Estudiantes por semestre
**Columna:** índice 2 (semestre).  
Se convierte a entero y se filtra para los semestres 5, 6 y 7.

```python
semestres = {5: 0, 6: 0, 7: 0}

for fila in filas:
    sem = int(fila[2].strip())
    if sem in semestres:
        semestres[sem] += 1

for sem, total in semestres.items():
    print(f"Semestre {sem}: {total}")
```

---
## Reporte 4 — Estudiantes por jornada
**Columna:** índice 3 (jornada).  
Se clasifica en Matutina, Vespertina y Nocturna.

```python
jornadas = ["Matutina", "Vespertina", "Nocturna"]
conteo_jornadas = {j: 0 for j in jornadas}

for fila in filas:
    jornada = fila[3].strip()
    if jornada in conteo_jornadas:
        conteo_jornadas[jornada] += 1

for jornada, total in conteo_jornadas.items():
    print(f"{jornada}: {total}")
```

## Reporte 5 — Estudiantes que trabajan
**Columna:** índice 4 (trabaja).  
Se separa entre quienes respondieron Sí y quienes respondieron No, calculando el porcentaje de cada grupo.

```python
trabajan     = sum(1 for fila in filas if fila[4].strip() == "Sí")
no_trabajan  = sum(1 for fila in filas if fila[4].strip() == "No")
total        = trabajan + no_trabajan

print(f"Trabajan    : {trabajan}  ({trabajan / total * 100:.2f}%)")
print(f"No trabajan : {no_trabajan}  ({no_trabajan / total * 100:.2f}%)")
```

# Reporte 14: Carrera Percibida como Más Difícil
Se analiza el promedio de dificultad reportado por los estudiantes (`q15_dificultad_cursos`), agrupándolo por la carrera a la que pertenecen.

```python
def generar_reporte_14():
    datos = procesar_datos_encuesta()
    if not datos: return
    
    stats_carreras = [
        ["Ingeniería de Sistemas", 0, 0],
        ["Ingeniería Electrónica", 0, 0],
        ["Ingeniería Eléctrica", 0, 0]
    ]
    
    for encuestado in datos:
        carrera = encuestado[1][0]
        dificultad = encuestado[5][0] # q15
        
        for i in range(len(stats_carreras)):
            if stats_carreras[i][0] == carrera:
                stats_carreras[i][1] += dificultad
                stats_carreras[i][2] += 1
                break
                
    max_promedio = 0
    carrera_top = ""
    
    print("--- Reporte 14: Carrera percibida como más difícil (Basado en q15) ---")
    for item in stats_carreras:
        if item[2] > 0:
            promedio = item[1] / item[2]
            print(f"{item[0]}: Nivel promedio de dificultad = {promedio:.2f}")
            if promedio > max_promedio:
                max_promedio = promedio
                carrera_top = item[0]
                
    print("-" * 65)
    print(f"RESULTADO: La carrera percibida como más difícil es '{carrera_top}' con un promedio de {max_promedio:.2f}")
```

# Reporte 15: Carrera con Mayor Nivel Promedio de Estrés
Cruza la información del nivel de estrés académico (`q16_estres_academico`) con las diferentes carreras de ingeniería para identificar la más demandante en este aspecto.

```python
def generar_reporte_15():
    datos = procesar_datos_encuesta()
    if not datos: return
    
    stats_estres = [
        ["Ingeniería de Sistemas", 0, 0],
        ["Ingeniería Electrónica", 0, 0],
        ["Ingeniería Eléctrica", 0, 0]
    ]
    
    for encuestado in datos:
        carrera = encuestado[1][0]
        estres = encuestado[5][1] # q16
        
        for i in range(len(stats_estres)):
            if stats_estres[i][0] == carrera:
                stats_estres[i][1] += estres
                stats_estres[i][2] += 1
                break
                
    max_promedio = 0
    carrera_top = ""
    
    print("--- Reporte 15: Carrera con mayor nivel promedio de estrés (Basado en q16) ---")
    for item in stats_estres:
        if item[2] > 0:
            promedio = item[1] / item[2]
            print(f"{item[0]}: Nivel promedio de estrés = {promedio:.2f}")
            if promedio > max_promedio:
                max_promedio = promedio
                carrera_top = item[0]
                
    print("-" * 70)
    print(f"RESULTADO: La carrera con mayor estrés es '{carrera_top}' con un promedio de {max_promedio:.2f}")
```

## Reporte 16 — Relación entre trabajar y promedio académico. 

En este analisis se utilizo una estructura de listas anidadas para organizar mejor la informacion de cada estudiante

Cada fila del archivo CSV representa a un estudiante se reorganizan los datos en una estructura:

- Posicion 0 = ID del estudiante  
- Posicion 1 = Datos generales: carrera, semestre y jornada 
- Posicion 2 = Datos academicos: trabaja (Sí o No) y promedio academico
- Posicion 3 = Habitos de estudio: horas de estudio y planificacion del estudio
- Posicion 4 = Otros datos:
  - q03_entrega_tareas  
  - q04_trabajo_equipo 
  - q05_asistencia 
  - q06_uso_biblioteca
  - q07_uso_plataformas  
  - q08_consulta_tutorias  
  - q09_disponibilidad_laboratorios  
  - q10_calidad_internet  
  - q11_claridad_docente  
  - q12_puntualidad_docente  
  - q13_retroalimentacion  
  - q14_evaluacion_justa  
  - q15_dificultad_cursos  
  - q16_estres_academico  
  - q17_equilibrio_vida_estudio  
  - q18_satisfaccion_carrera 
  - q19_intencion_continuar  
  - q20_recomendaria_facultad  
Estos datos se agrupan automaticamente lo que permite incluir todas las respuestas restantes sin tener que escribirlas una por una 
Por ejemplo:
- `estudiante[2][0]` permite saber si el estudiante trabaja  
- `estudiante[2][1]` permite obtener su promedio  

```python
encuesta = []

for fila in filas:
    estudiante = [
        fila[0],
        [fila[1], fila[2], fila[3]],
        [fila[4], fila[5]],
        [fila[6], fila[7]],
        fila[8:]
    ]
    encuesta.append(estudiante)
```

```python
suma_si = 0
contador_si = 0

suma_no = 0
contador_no = 0

for estudiante in encuesta:
    trabaja = estudiante[2][0]
    promedio = float(estudiante[2][1])

    if trabaja == "Sí":
        suma_si += promedio
        contador_si += 1
    elif trabaja == "No":
        suma_no += promedio
        contador_no += 1

promedio_si = suma_si / contador_si if contador_si > 0 else 0
promedio_no = suma_no / contador_no if contador_no > 0 else 0

print(f"Promedio estudiantes que trabajan: {promedio_si:.2f}")
print(f"Promedio estudiantes que no trabajan: {promedio_no:.2f}")
```

## Reporte 17 —  Relación entre acceso a internet y promedio académico.

Se analiza la relacion entre la calidad del acceso a internet y el rendimiento academico de los estudiantes comparando los promedios entre quienes tienen buen y mal acceso.

```python
# Iniciamos acumuladores y contadores
suma_bueno = 0
contador_bueno = 0

suma_malo = 0
contador_malo = 0

# Recorremos los estudiantes
for estudiante in encuesta:

    # Obtenemos el promedio academico
    promedio = float(estudiante[2][1])

    # La calidad de internet corresponde a q10_calidad_internet
    # Está dentro de "otros datos" (fila[8:])
    # q10 es la posición 7 dentro de estudiante[4]
    internet = int(estudiante[4][7])

    # Clasificamos segun la calidad de internet
    if internet >= 3:
        suma_bueno += promedio
        contador_bueno += 1
    else:
        suma_malo += promedio
        contador_malo += 1


# Calculo de promedios
if contador_bueno > 0:
    promedio_bueno = suma_bueno / contador_bueno
else:
    promedio_bueno = 0

if contador_malo > 0:
    promedio_malo = suma_malo / contador_malo
else:
    promedio_malo = 0


# Mostrar resultados
print("Promedio teniendo buen internet:", round(promedio_bueno, 2))
print("Promedio teniendo mal internet:", round(promedio_malo, 2))
```

## Reporte 18 —  Relación entre horas de estudio y rendimiento académico. 

Se analiza como la cantidad de horas de estudio influye en el rendimiento academico comparando los promedios segun distintos niveles de dedicacion que le dan al estudio

```python
# Acumuladores y contadores
suma_1 = 0
cont_1 = 0

suma_2 = 0
cont_2 = 0

suma_3 = 0
cont_3 = 0

suma_4 = 0
cont_4 = 0

suma_5 = 0
cont_5 = 0

# Recorrido de estudiantes
for estudiante in encuesta:

    horas = int(estudiante[3][0])        # q01_horas_estudio
    promedio = float(estudiante[2][1])  # promedio academico

    if horas == 1:
        suma_1 += promedio
        cont_1 += 1

    elif horas == 2:
        suma_2 += promedio
        cont_2 += 1

    elif horas == 3:
        suma_3 += promedio
        cont_3 += 1

    elif horas == 4:
        suma_4 += promedio
        cont_4 += 1

    elif horas == 5:
        suma_5 += promedio
        cont_5 += 1

# Calculo de promedios
prom_1 = suma_1 / cont_1 if cont_1 > 0 else 0
prom_2 = suma_2 / cont_2 if cont_2 > 0 else 0
prom_3 = suma_3 / cont_3 if cont_3 > 0 else 0
prom_4 = suma_4 / cont_4 if cont_4 > 0 else 0
prom_5 = suma_5 / cont_5 if cont_5 > 0 else 0

# Mostrar resultados
print("Promedio con 1 hora de estudio:", round(prom_1, 2))
print("Promedio con 2 horas de estudio:", round(prom_2, 2))
print("Promedio con 3 horas de estudio:", round(prom_3, 2))
print("Promedio con 4 horas de estudio:", round(prom_4, 2))
print("Promedio con 5 horas de estudio:", round(prom_5, 2)) 

# Se determina cual tiene el mejor promedio
mejor_promedio = prom_1
mejor_horas = 1

if prom_2 > mejor_promedio:
    mejor_promedio = prom_2
    mejor_horas = 2

if prom_3 > mejor_promedio:
    mejor_promedio = prom_3
    mejor_horas = 3

if prom_4 > mejor_promedio:
    mejor_promedio = prom_4
    mejor_horas = 4

if prom_5 > mejor_promedio:
    mejor_promedio = prom_5
    mejor_horas = 5

print("El mejor promedio lo obtienen los estudiantes que estudian", mejor_horas, "horas")
```

## Reporte 19 — Porcentaje de estudiantes interesados en cursos virtuales.   

Se calcula el porcentaje de estudiantes interesados en cursos virtuales a partir de las respuestas de la encuesta

```python
# Iniciamos contadores
total = 0
interesados = 0

# Recorremos los estudiantes
for estudiante in encuesta:

    total += 1  # Contamos todos los estudiantes

    # Se obtiene la respuesta q19 desde "otros"
    interes = int(estudiante[4][16]) # La posición [16] corresponde a q19 dentro del bloque "otros"

    # Se consideran interesados los valores 4 y 5
    if interes >= 4:
        interesados += 1
# Calculo del porcentaje
if total > 0:
    porcentaje = (interesados / total) * 100
else:
    porcentaje = 0

# Mostrar resultado
print("Porcentaje de estudiantes interesados en cursos virtuales:", round(porcentaje, 2), "%")
```
