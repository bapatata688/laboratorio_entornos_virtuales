# Analisis de Estudiantes por Jornada

Se lee la columna `N4` (índice 3) del CSV para contar cuántos estudiantes pertenecen a cada jornada: Matutina, Vespertina y Nocturna.

## Inicialización de variables

Se define una lista de referencia con los tres valores posibles de jornada, un contador por cada una y una lista anidada donde cada sublista acumulará los conteos de su jornada correspondiente.

```python
import csv

jornada = ["Matutina", "Vespertina", "Nocturna"]

turno_matutino   = 0
turno_vespertino = 0
turno_nocturno   = 0

# jornadas[0] → Matutina | jornadas[1] → Vespertina | jornadas[2] → Nocturna
jornadas = [[], [], []]
```

## Lectura del CSV

`next(encuesta)` descarta la fila de encabezados. Por cada fila se lee `turno[3]` (columna N4) y se compara con cada jornada. Al haber coincidencia, el contador sube y su valor se agrega a la sublista correspondiente.

```python
with open("../encuesta_ingenieria_10000_respuestas.csv", "r", encoding="utf-8") as encuestas:
    encuesta = csv.reader(encuestas)
    next(encuesta)

    for turno in encuesta:
        turno_jornada = turno[3].strip()

        if turno_jornada == jornada[0]:
            turno_matutino += 1
            jornadas[0].append(turno_matutino)

        if turno_jornada == jornada[1]:
            turno_vespertino += 1
            jornadas[1].append(turno_vespertino)

        if turno_jornada == jornada[2]:
            turno_nocturno += 1
            jornadas[2].append(turno_nocturno)
```

## Resultados

`len()` sobre cada sublista devuelve el total de estudiantes por jornada, ya que cada elemento representa un registro contado.

```python
total_turno_matutino   = len(jornadas[0])
total_turno_vespertino = len(jornadas[1])
total_turno_nocturno   = len(jornadas[2])

print(f"Matutino: {total_turno_matutino}  Vespertino: {total_turno_vespertino}  Nocturno: {total_turno_nocturno}")
```
