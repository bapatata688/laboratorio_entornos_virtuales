# Reporte 16  Relación entre trabajar y promedio académico. 

## Carga de datos
Primero se realiza la lectura del archivo CSV que contiene la informacion de la encuesta.

Se utiliza la librería CSV para leer el archivo y se omite la primera fila de los encabezados, luego cada fila que son las que representas a los estudiantes se almacenara en la lista principal llamada estudiante.

```python
# Importamos la libreria csv para leer el archivo
import csv

encuesta = []

with open("../encuesta_ingenieria_10000_respuestas.csv", encoding="utf-8") as archivo:

    lector = csv.reader(archivo)
    next(lector) # Omitimos la primera fila de los encabezados

    for fila in lector:

        estudiante = [
            fila[0],  # ID del encuestado

            [fila[1], fila[2], fila[3]],  # Datos generales carrera, semestre, jornada

            [fila[4], fila[5]],  # Datos academicos trabaja, promedio

            [fila[6], fila[7]],  # Habitos de estudio

            fila[8:]   # Otros datos de la encuesta
        ]

        encuesta.append(estudiante) # Agregando estudiante a la lista principal

print("Total de estudiantes cargados:", len(encuesta)) # Verificamos si todos los datos de las encuestas fueron cargados

```

## Procesamiento de datos
Clasificamos los dos grupos de estudiantes entre los que trabajan y los que no trabajan.

Para cada grupo se acumulan los promedios y se cuenta la cantidad de estudiantes para calcular el promedio de cada categoria.

```python
# Acumuladores y contadores
suma_si = 0
contador_si = 0

suma_no = 0
contador_no = 0

# Recorrido de los estudiantes
for estudiante in encuesta:

    # Se accede a los datos academicos
    trabaja = estudiante[2][0]
    promedio = float(estudiante[2][1])  # Se convierte a numero para poder operar y se pone un float para que acepte decimales

    # Se clasifican los que trabajan y los que no trabajan 
    if trabaja == "Sí":
        suma_si += promedio
        contador_si += 1

    elif trabaja == "No":
        suma_no += promedio
        contador_no += 1
```

## Resultados
Se calculan los promedios academicos para cada grupo y se muestran para  comparar el rendimiento entre estudiantes que trabajan y los que no.

```python
# Calculo de promedios
if contador_si > 0:
    promedio_si = suma_si / contador_si
else:
    promedio_si = 0

if contador_no > 0:
    promedio_no = suma_no / contador_no
else:
    promedio_no = 0

# Mostrar resultados
print("Promedio estudiantes que trabajan:", round(promedio_si, 2))
print("Promedio estudiantes que no trabajan:", round(promedio_no, 2))
```
