# Reporte 17 Relación entre acceso a internet y promedio académico. 

 ## Carga de datos
Primero se realiza la lectura del archivo CSV que contiene la informacion de la encuesta.

Se utiliza la librería CSV para leer el archivo y se omite la primera fila de los encabezados, luego cada fila que son las que representas a los estudiantes se almacenara en la lista principal llamada estudiante.

```python
# Importamos la libreria csv para leer el archivo
import csv

# Lista donde se guardaran los estudiantes
encuesta = []

# Abrimos el archivo CSV
with open("../encuesta_ingenieria_10000_respuestas.csv", encoding="utf-8") as archivo:

    lector = csv.reader(archivo)

    # Omitimos la fila de encabezados
    next(lector)

    # Recorremos cada fila 
    for fila in lector:

        # Organizamos los datos en listas anidadas
        estudiante = [
            fila[0],  # ID del encuestado

            [fila[1], fila[2], fila[3]],  # Datos generales carrera, semestre, jornada

            [fila[4], fila[5]],  # Datos academicos trabaja, promedio

            [fila[6], fila[7]],  # Habitos de estudio

            fila[8:]  # Otros datos incluyendo el internet
        ]

        # Guardamos el estudiante en la lista principal
        encuesta.append(estudiante)

# Verificamos la carga de datos
print("Total de estudiantes cargados:", len(encuesta))
```

## Procesamiento de datos 
Se recorre la lista de estudiantes y se clasifica a cada uno segun la calidad de su acceso a internet.

Se considera "buen internet" a los valores mayores o iguales a 3 y "mal internet" a los valores menores a 3.

Para cada grupo se acumulan los promedios y se cuenta la cantidad de estudiantes.

```python
# Iniciamos acumuladores y contadores
suma_bueno = 0
contador_bueno = 0

suma_malo = 0
contador_malo = 0

# Recorremos los estudiantes
for estudiante in encuesta:

    # Obtenemos el promedio academico
    promedio = float(estudiante[2][1])

    # La calidad de internet está en "otros"
    internet = int(estudiante[4][7]) # Se pone un int porque la variable de internet es entera del 1 al 5

    # Clasificamos segun la calidad de internet
    if internet >= 3:
        suma_bueno += promedio
        contador_bueno += 1
    else:
        suma_malo += promedio
        contador_malo += 1
```

## Resultado

Se calculan los promedios academicos para los estudiantes con buen acceso a internet y con mal acceso para comparar el rendimiento entre ambos grupos.

```python
# Calculo de promedios
if contador_bueno > 0:
    promedio_bueno = suma_bueno / contador_bueno
else:
    promedio_bueno = 0

if contador_malo > 0:
    promedio_malo = suma_malo / contador_malo
else:
    promedio_malo = 0

# Mostrar resultados
print("Promedio teniendo buen internet:", round(promedio_bueno, 2))
print("Promedio teniendo mal internet:", round(promedio_malo, 2))
```
