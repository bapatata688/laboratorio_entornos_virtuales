# ¿Qué es un entorno virtual?

## Definición

Un entorno virtual es un espacio aislado dentro de tu sistema donde puedes instalar y gestionar paquetes de Python de forma independiente.

## ¿Para qué sirve?

Permite que cada proyecto tenga sus propias:

- librerías
- versiones de paquetes
- configuraciones

## Problema que resuelve

Evita conflictos entre proyectos que usan diferentes versiones de una misma librería.

## Ejemplo conceptual

Proyecto A:
- pandas 1.0

Proyecto B:
- pandas 2.0

Sin entorno virtual → conflicto  
Con entorno virtual → ambos funcionan correctamente

## Idea clave

Un entorno virtual permite que cada proyecto tenga su propio ecosistema de dependencias.

## Crear un entorno virtual con venv

* python -m venv mi_entorno

Activar el entorno (Mac/Linux)

* source mi_entorno/bin/activate

Activar el entorno (Windows)

* mi_entorno\Scripts\activate

Instalar un paquete dentro del entorno

* pip install pandas

Ver paquetes instalados

* pip list

Desactivar el entorno virtual

* deactivate

# ¿Por qué usar entornos virtuales?

## Problema sin entornos virtuales

Cuando se instalan paquetes de forma global:

- todos los proyectos comparten las mismas librerías
- pueden existir conflictos de versiones
- es difícil mantener proyectos independientes

## Beneficios

- Aislamiento de proyectos
- Control de versiones de paquetes
- Reproducibilidad
- Organización profesional

## Ejemplo conceptual

Proyecto A necesita:
- pandas 1.5

Proyecto B necesita:
- pandas 2.0

Sin entorno virtual → conflicto  
Con entorno virtual → ambos funcionan correctamente

## Aplicación real

En ingeniería de datos cada proyecto puede requerir:

- librerías distintas
- versiones específicas
- configuraciones diferentes

## Idea clave

Cada proyecto debe tener su propio entorno para evitar conflictos y garantizar estabilidad.

### Código

# Problemas de trabajar sin entornos virtuales

## ¿Qué sucede sin entornos virtuales?

Todos los proyectos usan las mismas librerías instaladas globalmente.

## Problemas principales

### 1. Conflictos de versiones
Un proyecto puede requerir una versión distinta de una librería.

### 2. Romper otros proyectos
Actualizar una librería puede hacer que otro proyecto deje de funcionar.

### 3. Dificultad para reproducir proyectos
Otro usuario no puede instalar exactamente las mismas dependencias fácilmente.

### 4. Desorden en el sistema
Se acumulan muchas librerías sin control.

## Ejemplo conceptual

Instalas:
- pandas 2.0

Pero otro proyecto usa:
- pandas 1.5

Resultado → errores

## Idea clave

Trabajar sin entornos virtuales hace que los proyectos sean inestables y difíciles de mantener.

# Instalación global vs entorno aislado

## Instalación global

Los paquetes se instalan en el sistema y son compartidos por todos los proyectos.

### Características
- disponible para todos los proyectos
- riesgo de conflictos
- difícil control de versiones

## Entorno aislado

Los paquetes se instalan solo dentro de un entorno específico.

### Características
- independiente por proyecto
- sin conflictos
- control total de versiones

## Comparación

Global:
- un solo conjunto de librerías

Aislado:
- un entorno por proyecto

## Idea clave

La instalación global es simple,  
pero el entorno aislado es profesional.

## Código 

Instalación global

* pip install pandas

Crear entorno aislado

* python -m venv entorno

Activar entorno

* source entorno/bin/activate

Instalar paquete dentro del entorno

* pip install pandas

Ver paquetes instalados (solo en el entorno)

* pip list

Salir del entorno

* deactivate

- pip
- conda

Facilita la gestión de dependencias en un proyecto.

## Funciones principales

- instalar paquetes

- resolver dependencias automáticamente

Un gestor de paquetes automatiza la instalación y manejo de librerías.
