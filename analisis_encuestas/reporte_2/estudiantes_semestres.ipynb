# Analisis de Estudiantes por Semestre

Se lee la columna de semestre (índice 2) del CSV para contar estudiantes en los semestres 5, 6 y 7.

## Inicialización de variables

Se definen los semestres como enteros de referencia, un contador por cada uno y una lista anidada donde cada sublista acumula los registros correspondientes.

```python
import csv

semestre_5 = 5
semestre_6 = 6
semestre_7 = 7

estudiantes_semestre_5 = 0
estudiantes_semestre_6 = 0
estudiantes_semestre_7 = 0

# [0] → semestre 5 | [1] → semestre 6 | [2] → semestre 7
estudiante_semestre = [[], [], []]
```

## Lectura del CSV

`semestre[2]` es la tercera columna (índice 2). Se convierte a entero con `int()` para poder compararlo con los valores de referencia.

> **Nota:** la tercera condicion del bucle compara contra `semestre_6` en lugar de `semestre_7`, lo que hace que `estudiante_semestre[2]` cuente los mismos registros que `[1]`. Corregir esa línea a `semestre_7` para obtener el conteo real del semestre 7.

```python
with open("../encuesta_ingenieria_10000_respuestas.csv", "r", encoding="utf-8") as estudiantes:
    encuesta = csv.reader(estudiantes)
    next(encuesta)

    for semestre in encuesta:
        estudiantes_semestre = int(semestre[2].strip())

        if estudiantes_semestre == semestre_5:
            estudiantes_semestre_5 += 1
            estudiante_semestre[0].append(estudiantes_semestre_5)

        if estudiantes_semestre == semestre_6:
            estudiantes_semestre_6 += 1
            estudiante_semestre[1].append(estudiantes_semestre_6)

        if estudiantes_semestre == semestre_6:   # corregir a semestre_7
            estudiantes_semestre_7 += 1
            estudiante_semestre[2].append(estudiantes_semestre_7)
```

## Resultados

`len()` sobre cada sublista devuelve el total de estudiantes por semestre.

```python
total_semestre_5 = len(estudiante_semestre[0])
total_semestre_6 = len(estudiante_semestre[1])
total_semestre_7 = len(estudiante_semestre[2])

print(f"Semestre 5: {total_semestre_5}")
print(f"Semestre 6: {total_semestre_6}")
print(f"Semestre 7: {total_semestre_7}")
```
