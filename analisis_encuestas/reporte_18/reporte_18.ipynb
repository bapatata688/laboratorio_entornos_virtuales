# Reporte 18 Relación entre horas de estudio y rendimiento académico. 

## Carga de datos
Primero se realiza la lectura del archivo CSV que contiene la informacion de la encuesta.

Se utiliza la librería CSV para leer el archivo y se omite la primera fila de los encabezados, luego cada fila que son las que representas a los estudiantes se almacenara en la lista principal llamada estudiante.

```python
import csv

encuesta = []

with open("../encuesta_ingenieria_10000_respuestas.csv", encoding="utf-8") as archivo:
    lector = csv.reader(archivo)
    next(lector)

    for fila in lector:

        estudiante = [
            fila[0],  # ID

            [fila[1], fila[2], fila[3]],  # datos generales

            [fila[4], fila[5]],  # datos academicos

            [fila[6], fila[7]],  # Habitos de estudio 

            fila[8:]  # otros
        ]

        encuesta.append(estudiante)

print("Total de estudiantes cargados:", len(encuesta))
```

## Procesamiento de datos
Se recorre la lista de estudiantes y se clasifican segun la cantidad de horas de estudio semanal (valores de 1 a 5).

Para cada grupo se acumulan los promedios academicos y se cuenta la cantidad de estudiantes para el promedio por cada nivel de estudio.

```python
# Acumuladores y contadores
suma_1 = 0
cont_1 = 0

suma_2 = 0
cont_2 = 0

suma_3 = 0
cont_3 = 0

suma_4 = 0
cont_4 = 0

suma_5 = 0
cont_5 = 0

# Recorrido de estudiantes
for estudiante in encuesta:

    horas = int(estudiante[3][0])        # q01_horas_estudio
    promedio = float(estudiante[2][1])  # promedio academico

    if horas == 1:
        suma_1 += promedio
        cont_1 += 1

    elif horas == 2:
        suma_2 += promedio
        cont_2 += 1

    elif horas == 3:
        suma_3 += promedio
        cont_3 += 1

    elif horas == 4:
        suma_4 += promedio
        cont_4 += 1

    elif horas == 5:
        suma_5 += promedio
        cont_5 += 1
```

## Resultado

Se calculan los promedios academicos para cada nivel de horas de estudio y se muestran en pantalla para comparar el rendimiento.

```python
# Calculo de promedios
prom_1 = suma_1 / cont_1 if cont_1 > 0 else 0
prom_2 = suma_2 / cont_2 if cont_2 > 0 else 0
prom_3 = suma_3 / cont_3 if cont_3 > 0 else 0
prom_4 = suma_4 / cont_4 if cont_4 > 0 else 0
prom_5 = suma_5 / cont_5 if cont_5 > 0 else 0

# Mostrar resultados
print("Promedio con 1 hora de estudio:", round(prom_1, 2))
print("Promedio con 2 horas de estudio:", round(prom_2, 2))
print("Promedio con 3 horas de estudio:", round(prom_3, 2))
print("Promedio con 4 horas de estudio:", round(prom_4, 2))
print("Promedio con 5 horas de estudio:", round(prom_5, 2))
```

```python
Se identifica cual grupo de horas de estudio obtiene el mayor promedio academico.
```

```python
# Se determina cual tiene el mejor promedio
mejor_promedio = prom_1
mejor_horas = 1

if prom_2 > mejor_promedio:
    mejor_promedio = prom_2
    mejor_horas = 2

if prom_3 > mejor_promedio:
    mejor_promedio = prom_3
    mejor_horas = 3

if prom_4 > mejor_promedio:
    mejor_promedio = prom_4
    mejor_horas = 4

if prom_5 > mejor_promedio:
    mejor_promedio = prom_5
    mejor_horas = 5

print("El mejor promedio lo obtienen los estudiantes que estudian", mejor_horas, "horas")
```
