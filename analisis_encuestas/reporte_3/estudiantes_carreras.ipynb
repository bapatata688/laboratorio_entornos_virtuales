# Analisis de Estudiantes por Carrera

Se lee la columna de carreras (índice 1) del CSV para contar cuántos estudiantes pertenecen a cada ingeniería.

## Inicialización de variables

Se define la lista de referencia con los tres valores posibles de carrera y una lista anidada donde cada sublista acumula registros de su carrera.

```python
import csv

carreras = ["Ingeniería Eléctrica", "Ingeniería de Sistemas", "Ingeniería Electrónica"]

ing_electrica   = 0
ing_sistemas    = 0
ing_electronica = 0

# [0] → Sistemas | [1] → Eléctrica | [2] → Electrónica
estudiantes_ingienerias = [[], [], []]
```

## Lectura del CSV

`encuestado[1]` corresponde a la segunda columna del CSV (índice 1), que contiene la carrera de cada estudiante. Según la coincidencia, el registro se agrega a la sublista correspondiente.

```python
with open("../encuesta_ingenieria_10000_respuestas.csv", "r", encoding="utf-8") as estudiantes:
    encuesta = csv.reader(estudiantes)
    next(encuesta)

    for encuestado in encuesta:
        carrera = encuestado[1]

        if carrera == carreras[0]:
            estudiantes_ingienerias[1].append(ing_electrica)
        elif carrera == carreras[1]:
            estudiantes_ingienerias[0].append(ing_sistemas)
        elif carrera == carreras[2]:
            estudiantes_ingienerias[2].append(ing_electronica)
```

## Resultados

`len()` sobre cada sublista devuelve el total de estudiantes registrados por carrera.

```python
total_estudiantes_electricos   = len(estudiantes_ingienerias[0])
total_estudiantes_sistemas     = len(estudiantes_ingienerias[1])
total_estudiantes_electronicos = len(estudiantes_ingienerias[2])

print(f"Ingeniería Eléctrica  : {total_estudiantes_electricos}")
print(f"Ingeniería de Sistemas: {total_estudiantes_sistemas}")
print(f"Ingeniería Electrónica: {total_estudiantes_electronicos}")
```
