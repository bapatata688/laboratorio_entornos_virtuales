#  Análisis de Estudiantes Trabajadores

Este notebook analiza una encuesta de ingeniería para determinar **qué porcentaje de estudiantes trabaja** y cuál no, leyendo los datos desde un archivo CSV.

---

## 1. Importación de librerías

Usamos el módulo `csv` de la librería estándar de Python para leer archivos con valores separados por comas.

````python
import csv
````

## 2. Inicialización de variables

Antes de leer el CSV, preparamos las estructuras que vamos a necesitar:

| Variable | Tipo | Propósito |
|---|---|---|
| `estudiante_trabaja` | Lista | Valores de referencia: `'Sí'` y `'No'` |
| `trabajador` | Entero | Contador acumulado de estudiantes que trabajan |
| `no_trabajador` | Entero | Contador acumulado de estudiantes que no trabajan |
| `estudiantes` | Lista anidada | Guarda los conteos separados por grupo |

````python
# Valores posibles en la columna de empleo
estudiante_trabaja = ["Sí", "No"]

# Contadores independientes para cada grupo
trabajador = 0
no_trabajador = 0

# Lista anidada: índice 0 → trabajadores | índice 1 → no trabajadores
estudiantes = [[], []]
````

## 3. Lectura del archivo CSV

Abrimos el archivo con `open()` usando:
- `'r'` → modo lectura.
- `encoding='utf-8'` → para soportar caracteres especiales como tildes.

Luego usamos `csv.reader` para iterar fila por fila.

>  **Columna analizada:** `trabajadores[4]` corresponde a la **quinta columna** (índice 4), llamada `N5` en la encuesta, que registra si el estudiante trabaja o no.

````python
with open(
    "../encuesta_ingenieria_10000_respuestas.csv", "r", encoding="utf-8"
) as estudiante:

    encuesta = csv.reader(estudiante)  # Objeto lector del CSV
    next(encuesta)                     # Salta la fila de encabezados

    for trabajadores in encuesta:
        # Extraemos la columna N5 y eliminamos espacios en blanco
        trabaja = trabajadores[4].strip()

        # Si trabaja → aumentar contador y agregar a la lista [0]
        if trabaja == estudiante_trabaja[0]:   # "Sí"
            trabajador += 1
            estudiantes[0].append(trabajador)

        # Si no trabaja → aumentar contador y agregar a la lista [1]
        if trabaja == estudiante_trabaja[1]:   # "No"
            no_trabajador += 1
            estudiantes[1].append(no_trabajador)
````

## 4. Cálculo de totales y porcentajes

Con `len()` obtenemos cuántos elementos acumuló cada sublista, lo que equivale al total de cada grupo.

La fórmula del porcentaje es:

$$\text{Porcentaje} = \frac{\text{Estudiantes del grupo}}{\text{Total de estudiantes}} \times 100$$

````python
# Total de registros válidos procesados
total_trabajador = len(estudiantes[0]) + len(estudiantes[1])

# Totales por grupo
estudiantes_trabajadores     = len(estudiantes[0])
estudiantes_no_trabajadores  = len(estudiantes[1])

# Porcentajes
porcentaje_estudiantes_trabajadores = (
    estudiantes_trabajadores / total_trabajador
) * 100

porcentaje_estudiantes_no_trabajadores = (
    estudiantes_no_trabajadores / total_trabajador
) * 100
````

## 5. Resultados

Mostramos los porcentajes con dos decimales usando f-strings y el formato `:.2f`.

````python
print(f"Total de estudiantes analizados : {total_trabajador}")
print(f"Estudiantes que trabajan        : {estudiantes_trabajadores}")
print(f"Estudiantes que no trabajan     : {estudiantes_no_trabajadores}")
print()
print(f"Porcentaje de estudiantes que trabajan    : {porcentaje_estudiantes_trabajadores:.2f}%")
print(f"Porcentaje de estudiantes que no trabajan : {porcentaje_estudiantes_no_trabajadores:.2f}%")
````

---

##  Resumen del flujo

```
CSV  →  csv.reader  →  Bucle fila a fila
                              │
                 ┌────────────┴────────────┐
            trabaja == "Sí"         trabaja == "No"
                 │                         │
        estudiantes[0]            estudiantes[1]
                 └────────────┬────────────┘
                              │
                    len() → porcentajes
```

> **Nota:** El uso de `strip()` es importante para evitar errores de comparación por espacios invisibles en los datos del CSV.
