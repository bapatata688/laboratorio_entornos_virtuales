# Conteo Total de Encuestados

Se lee la primera columna (índice 0) del CSV, que contiene el ID de cada encuestado, para obtener el total de registros.

## Inicialización de variables

`encuestados` almacenará cada ID leído. `encuestado` es un contador auxiliar que incrementa por cada fila válida.

```python
import csv

encuestados = []
encuestado  = 0
```

## Lectura del CSV

`encuesta[0]` corresponde a la primera columna, el ID del encuestado. El `if id_encuestado` evita agregar filas vacías. El contador `encuestado` sube junto con cada `append`, aunque el total final se obtiene con `len()` sobre la lista.

```python
with open("../encuesta_ingenieria_10000_respuestas.csv", "r", encoding="utf-8") as encuestas:
    encuesta = csv.reader(encuestas)
    next(encuesta)

    for encuesta in encuesta:
        id_encuestado = encuesta[0]

        if id_encuestado:
            encuestados.append(id_encuestado)
            encuestado += 1
```

## Resultado

```python
total_encuestados = len(encuestados)
print(f"Total de encuestados: {total_encuestados}")
```
