# Reporte 7 — Promedio Académico por Carrera

Se agrupan los valores de `promedio_actual` (índice 5) por `carrera` (índice 1) y se calcula la media de cada grupo.

## Inicialización de variables

`prom_carrera` es una lista anidada donde cada elemento tiene la forma `[nombre_carrera, suma_promedios, cantidad]`. Esto permite calcular el promedio al final sin almacenar todos los valores individuales.

```python
import csv

# Columnas: id(0) carrera(1) semestre(2) jornada(3) trabaja(4) promedio_actual(5)
INDICE_CARRERA  = 1
INDICE_PROMEDIO = 5

# [nombre_carrera, suma_promedios, cantidad_estudiantes]
prom_carrera = []
```

## Lectura del CSV

Por cada fila se busca si la carrera ya existe en `prom_carrera`. Si existe, se acumula el promedio y el contador. Si no, se agrega una nueva entrada con `encontrado = False`.

```python
with open("../encuesta_ingenieria_10000_respuestas.csv", "r", encoding="utf-8") as estudiante:
    encuesta = csv.reader(estudiante)
    next(encuesta)

    for fila in encuesta:
        carrera  = fila[INDICE_CARRERA].strip()
        promedio = float(fila[INDICE_PROMEDIO].strip())
        encontrado = False

        for c in prom_carrera:
            if c[0] == carrera:
                c[1] += promedio
                c[2] += 1
                encontrado = True
                break

        if not encontrado:
            prom_carrera.append([carrera, promedio, 1])
```

## Resultado

Se divide la suma acumulada entre la cantidad de estudiantes de cada carrera.

```python
for c in prom_carrera:
    prom = c[1] / c[2]
    print(f"{c[0]}: {round(prom, 2)}")
```
