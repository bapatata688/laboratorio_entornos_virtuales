# Reporte 19 Porcentaje de estudiantes interesados en cursos virtuales.

## Carga de datos
Primero se realiza la lectura del archivo CSV que contiene la informacion de la encuesta.

Se utiliza la librería CSV para leer el archivo y se omite la primera fila de los encabezados, luego cada fila que son las que representas a los estudiantes se almacenara en la lista principal llamada estudiante.

```python
import csv

encuesta = []

with open("../encuesta_ingenieria_10000_respuestas.csv", encoding="utf-8") as archivo:
    lector = csv.reader(archivo)
    next(lector)

    for fila in lector:

        estudiante = [
            fila[0],  # ID

            [fila[1], fila[2], fila[3]],  # datos generales

            [fila[4], fila[5]],  # datos academicos

            [fila[6], fila[7]],  # Habitos de estudio

            fila[8:]  # otros 
        ]

        encuesta.append(estudiante)

print("Total de estudiantes cargados:", len(encuesta))
```

## Procesamiento de datos

Se recorre la lista de estudiantes y se evaluaa la respuesta correspondiente al interes en continuar cursos virtuales.

Se consideran como interesados los estudiantes que su respuesta sea mayor o igual a 4 en una escala de 1 a 5.

```python
# Iniciamos contadores
total = 0
interesados = 0

# Recorremos los estudiantes
for estudiante in encuesta:

    total += 1  # Contamos todos los estudiantes

    # Se obtiene la respuesta q19 desde "otros"
    interes = int(estudiante[4][16]) # La posición [16] corresponde a q19 dentro del bloque "otros"

    # Se consideran interesados los valores 4 y 5
    if interes >= 4:
        interesados += 1
```

## Resultado

Se calcula el porcentaje de estudiantes interesados en cursos virtuales en relación con el total de encuestados.

```python
# Calculo del porcentaje
if total > 0:
    porcentaje = (interesados / total) * 100
else:
    porcentaje = 0

# Mostrar resultado
print("Porcentaje de estudiantes interesados en cursos virtuales:", round(porcentaje, 2), "%")
```
