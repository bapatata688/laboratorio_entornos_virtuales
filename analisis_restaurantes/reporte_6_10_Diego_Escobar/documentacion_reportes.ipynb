# Documentacion: Analisis de Encuestas de Restaurante
Este notebook documenta el proceso de lectura de datos desde un archivo CSV y la generacion de reportes de analisis sobre las encuestas de un restaurante.

## 1. Estructura de datos

Se utiliza un diccionario llamado `restaurante` para almacenar todos los datos del CSV organizados por categoria.

| Clave principal | Sub-clave | Tipo | Descripcion |
|---|---|---|---|
| `id` | - | `list[int]` | Identificador unico por encuestado |
| `preferencias` | `comida` | `list[str]` | Tipo de comida preferida |
| `preferencias` | `frecuencia` | `list[str]` | Frecuencia de consumo (ej: Semanal, Diario) |
| `consumo` | `gasto` | `list[float]` | Gasto promedio por visita |
| `experiencia` | `producto` | `list[int]` | Calificacion de satisfaccion del producto (1-10) |
| `experiencia` | `servicio` | `list[int]` | Calificacion de satisfaccion del servicio (1-10) |
| `experiencia` | `tiempo` | `list[str]` | Percepcion del tiempo de entrega (Rapido, Aceptable, Lento) |
| `experiencia` | `precio` | `list[str]` | Percepcion del precio (Alto, Medio, Bajo) |
| `nps` | `recomendacion` | `list[int]` | Puntuacion de recomendacion (0-10) |
| `nps` | `volveria` | `list[bool]` | Si el cliente volveria (True/False) |
| `nps` | `general` | `list[int]` | Calificacion general (0-10) |

## 2. Lectura del CSV

Se usa `csv.DictReader` para leer el archivo. Cada fila del CSV se mapea directamente a las claves del diccionario `restaurante`. Los valores se convierten al tipo correspondiente durante la lectura.

Columnas del CSV: `id, comida_preferida, frecuencia_consumo, gasto_promedio, satisfaccion_producto, satisfaccion_servicio, tiempo_entrega, precio_percepcion, recomendaria, volveria_comprar, calificacion_general`

```python
import csv

restaurante = {
    "id": [],
    "preferencias": {"comida": [], "frecuencia": []},
    "consumo": {"gasto": []},
    "experiencia": {
        "producto": [],
        "servicio": [],
        "tiempo": [],
        "precio": [],
    },
    "nps": {
        "recomendacion": [],
        "volveria": [],
        "general": [],
    },
}

separador_de_reportes = "-" * 108

with open("encuesta_restaurantes_10000.csv", "r") as encuestas:
    datos_restaurante = csv.DictReader(encuestas)
    for valor in datos_restaurante:
        restaurante["id"].append(int(valor["id"]))
        restaurante["preferencias"]["comida"].append(valor["comida_preferida"])
        restaurante["preferencias"]["frecuencia"].append(valor["frecuencia_consumo"])
        restaurante["consumo"]["gasto"].append(float(valor["gasto_promedio"]))
        restaurante["experiencia"]["producto"].append(int(valor["satisfaccion_producto"]))
        restaurante["experiencia"]["servicio"].append(int(valor["satisfaccion_servicio"]))
        restaurante["experiencia"]["tiempo"].append(valor["tiempo_entrega"])
        restaurante["experiencia"]["precio"].append(valor["precio_percepcion"])
        restaurante["nps"]["recomendacion"].append(int(valor["recomendaria"]))
        restaurante["nps"]["volveria"].append(valor["volveria_comprar"].strip().lower() == "true")
        restaurante["nps"]["general"].append(int(valor["calificacion_general"]))
```

## 3. Reportes

Cada funcion de reporte accede directamente al diccionario global `restaurante`. No reciben argumentos.

### Reporte 6: Distribucion del tiempo de entrega

Cuenta cuantos pedidos fueron clasificados como Rapido, Aceptable o Lento, y calcula el porcentaje de cada uno sobre el total.

```python
def reporte_6():
    print(f"Reporte 6: distribucion del tiempo de entrega de pedidos")
    print(separador_de_reportes)
    tiempos_entrega = {"Rapido": 0, "Aceptable": 0, "Lento": 0}
    for conteo in restaurante["experiencia"]["tiempo"]:
        if conteo in tiempos_entrega:
            tiempos_entrega[conteo] += 1
    pedidos_rapido = tiempos_entrega["Rapido"]
    pedidos_aceptable = tiempos_entrega["Aceptable"]
    pedidos_lento = tiempos_entrega["Lento"]
    total_pedidos = pedidos_aceptable + pedidos_rapido + pedidos_lento
    porcentaje_pedidos_rapidos = (pedidos_rapido / total_pedidos) * 100
    porcentaje_pedidos_aceptables = (pedidos_aceptable / total_pedidos) * 100
    porcentaje_pedidos_lentos = (pedidos_lento / total_pedidos) * 100
    print(f"""
    total pedidos rapidos: {pedidos_rapido},
    total pedidos aceptables: {pedidos_aceptable},
    total pedidos lentos: {pedidos_lento}
    """)
    print(f"""
    porcentaje pedidos rapidos: {porcentaje_pedidos_rapidos}%,
    porcentaje pedidos aceptables: {porcentaje_pedidos_aceptables}%,
    porcentaje pedidos lentos: {porcentaje_pedidos_lentos}%
    """)
    print(separador_de_reportes)

reporte_6()
```

### Reporte 7: Distribucion de percepcion de precios

Cuenta cuantos clientes perciben el precio como Alto, Medio o Bajo, y calcula el porcentaje de cada categoria.

```python
def reporte_7():
    print("Reporte 7: distribucion de percepcion de precios")
    print(separador_de_reportes)
    percepcion_precios = {"Alto": 0, "Medio": 0, "Bajo": 0}
    experiencia_precio = restaurante["experiencia"]["precio"]
    for conteo in experiencia_precio:
        if conteo in percepcion_precios:
            percepcion_precios[conteo] += 1
    percepcion_precios_alto = percepcion_precios["Alto"]
    percepcion_precios_medio = percepcion_precios["Medio"]
    percepcion_precios_bajo = percepcion_precios["Bajo"]
    total_percepciones = percepcion_precios_alto + percepcion_precios_medio + percepcion_precios_bajo
    porcentaje_precio_alto = (percepcion_precios_alto / total_percepciones) * 100
    porcentaje_precio_medio = (percepcion_precios_medio / total_percepciones) * 100
    porcentaje_precio_bajo = (percepcion_precios_bajo / total_percepciones) * 100
    print(f"""
    personas que perciben precios altos: {percepcion_precios_alto}, 
    personas que perciben precios medios: {percepcion_precios_medio},
    personas que perciben precios bajos: {percepcion_precios_bajo}
    """)
    print(f"""
    porcentaje de personas que perciben precios altos: {porcentaje_precio_alto}%,
    porcentaje de personas que perciben precios medios: {porcentaje_precio_medio}%,
    porcentaje de personas que perciben precios bajos: {porcentaje_precio_bajo}%
    """)
    print(separador_de_reportes)

reporte_7()
```

### Reporte 8: Promedio de satisfaccion general

Calcula el promedio de la columna `recomendaria` como indicador de satisfaccion general.

```python
def reporte_8():
    print("Reporte 8: promedio de satisfaccion general")
    print(separador_de_reportes)
    recomendacion = restaurante["nps"]["recomendacion"]
    print(f"""
    promedio de satisfaccion general: {sum(recomendacion) / len(recomendacion):.2f}
    """)
    print(separador_de_reportes)

reporte_8()
```

### Reporte 9: Porcentaje de clientes que volverian

Cuenta cuantos clientes respondieron True o False en `volveria_comprar` y calcula el porcentaje de cada grupo.

```python
def reporte_9():
    print("Reporte 9: porcentaje de clientes que volverian")
    print(separador_de_reportes)
    clientes_volverian = restaurante["nps"]["volveria"]
    volverian = {"si": 0, "no": 0}
    for valor in clientes_volverian:
        if valor == True:
            volverian["si"] += 1
        if valor == False:
            volverian["no"] += 1
    clientes_que_volverian = volverian["si"]
    clientes_que_no_volverian = volverian["no"]
    total = clientes_que_no_volverian + clientes_que_volverian
    porcentaje_volverian = (clientes_que_volverian / total) * 100
    porcentaje_no_volverian = (clientes_que_no_volverian / total) * 100
    print(f""" 
    cantidad de clientes que volverian: {clientes_que_volverian}
    cantidad de clientes que no volverian: {clientes_que_no_volverian}
    porcentaje de clientes que regresarian: {porcentaje_volverian:.2f}%
    porcentaje de clientes que no regresarian: {porcentaje_no_volverian:.2f}
    """)
    print(separador_de_reportes)

reporte_9()
```

### Reporte 10: NPS (Net Promoter Score)

Clasifica a los encuestados segun su puntuacion en `recomendaria`:

| Grupo | Rango |
|---|---|
| Promotores | 9 - 10 |
| Pasivos | 7 - 8 |
| Detractores | 0 - 6 |

Formula: `NPS = %Promotores - %Detractores`

El resultado va de -100 a +100. Un valor positivo indica mas promotores que detractores.

```python
def reporte_10():
    print("Reporte 10: NPS (Net Promoter Score)")
    print(separador_de_reportes)
    recomendacion = restaurante["nps"]["recomendacion"]
    total = len(recomendacion)
    conteo = {"promotores": 0, "pasivos": 0, "detractores": 0}
    for recomendados in recomendacion:
        if recomendados >= 9:
            conteo["promotores"] += 1
        elif recomendados >= 7:
            conteo["pasivos"] += 1
        else:
            conteo["detractores"] += 1
    porcentajes = {
        "promotores": (conteo["promotores"] / total) * 100,
        "pasivos": (conteo["pasivos"] / total) * 100,
        "detractores": (conteo["detractores"] / total) * 100,
    }
    nps = porcentajes["promotores"] - porcentajes["detractores"]
    print(f"""
    promedio de NPS: {nps}
    """)
    print(separador_de_reportes)

reporte_10()
```
